## 2. Exploratory Data Analysis 

- Missing Data Handling

```python
# filling the null values of Review Rating column with the median of each category rating
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# confirming the changes
df.isnull().sum()
```

- Column Standardization

```python
# Column Standardization: modifying the column names to lower case separted with '_' to comply the data with SQL usage
df.columns = df.columns.str.replace(' ', '_').str.lower()
# Taking off the currncy unit to purchase_amount column
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# confirming the change and checking the rest of the columns names
df.columns
```

- Feature Engineering

```python
# create a new column to categorize the customers age.
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
# checking the fist 10 rows of the added column and the one that we used
df[['age','age_group']].head(10)
```

```python
#  create column purchase_freqency_days for categorization purposes of corresponding number of days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchase_freqency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# checking the new column
df[['frequency_of_purchases','purchase_freqency_days']].head(10)
```

- Data Consistency Check

```python
# checking discount_applied and promo_code_used
df[['discount_applied', 'promo_code_used']].head()
```

```python
# checking if the value in discount_applied is exactly the same as the value in promo_code_used
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# as result we will drop one of them to minimize the dataset
df = df.drop('promo_code_used',axis=1)
```
